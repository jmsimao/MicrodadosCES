#### Gráfico de Barras

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ces2020 = pd.read_csv('./dados/MICRODADOS_CADASTRO_CURSOS_2020_ALTO_TIETE.csv',
                   header=0,
                   sep=';',
                   encoding='latin-1',
                   low_memory=False
                  )
```

```python
cesGRU2020=ces2020[(ces2020['NO_MUNICIPIO']=='Guarulhos')&(ces2020['NO_CINE_AREA_GERAL']=='Educação')]
```

```python
cesGRU2020[['NO_CINE_AREA_GERAL','NO_CINE_AREA_DETALHADA',
            'QT_ING_BRANCA','QT_ING_PRETA','QT_ING_PARDA','QT_ING_AMARELA','QT_ING_INDIGENA','QT_ING_CORND']]
```

```python
cesGRU2020=cesGRU2020.groupby(by='NO_CINE_AREA_GERAL')[['QT_ING_BRANCA','QT_ING_PRETA','QT_ING_PARDA','QT_ING_AMARELA','QT_ING_INDIGENA','QT_ING_CORND']].sum()
cesGRU2020
```

```python
dados = []
etiquetas = ['BRANCA','PRETA','PARDA','AMARELA','INDIGENA','N/D']
cores=['red','green','blue','yellow','magenta','cyan']

for valor in cesGRU2020.values:
    dados.append(valor)



def barPlot(valores, etiquetasX, **kwargs):
    plt.figure(figsize=(8,10))
    plt.title(kwargs['titulo'])

    plt.bar(x=etiquetasX, height=valores, color=kwargs['cores'])


    plt.show()



barPlot(dados, etiquetas, 
        titulo='Gráfico 1',
        cores=cores
       )




```
